#Importação de bibliotecas e funções

```python
from sklearn.svm import OneClassSVM
from sklearn.linear_model import SGDOneClassSVM
from sklearn.covariance import EllipticEnvelope
from sklearn.metrics import confusion_matrix
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import classification_report
```

#Experimento 5 - CTU-13 (Captura 52)

```python
from sklearn.decomposition import FastICA


early_warning_url = "early_warning_exp_2.csv"
early_warning = pd.read_csv(early_warning_url, sep=";")
early_warning
colunas = early_warning.columns[361:]

df = early_warning[colunas]
df = df.fillna(0)
label = pd.read_csv('ctulabel.csv')['maior_2']

init = 97
train = 324
end = 778

x_train = df[init:train]
x_test = df[train:end]
y_real = label[train:end]

```

```python
pca = FastICA(n_components=3).fit(x_train)
novo_train  = pca.transform(x_train)
novo_test = []
for i in range (0,8):
    comeco = ((i*60))
    final = (((i+1)*60))
    if(final > end):
        final = end
    print(comeco,final)
    novo_test.append(pca.transform(x_test[comeco:final]))
    
array_kernel = ['linear' , 'poly' , 'rbf' , 'sigmoid' , ] # 'precomputed'
models = []
nu_atual = 0.01
for c in array_kernel:
    while nu_atual <= 0.5:
        models.append(OneClassSVM(gamma='auto', kernel=c, nu=nu_atual).fit(novo_train))
        nu_atual += 0.01
    nu_atual = 0.05

from sklearn.metrics import accuracy_score
for clf in models:
    preds = []
    for i in novo_test:
        y_test = clf.predict(i)
        y_test_final = [0 if i == 1 else 1 for i in y_test]
        preds += y_test_final
    if(accuracy_score(y_real, preds) > 0.83):
        print(clf)
        print(confusion_matrix(y_real, preds))
        print(classification_report(y_real, preds, digits = 4))
```
